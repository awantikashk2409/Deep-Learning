# Sigmoid (0 to 1)

```python
import math

def sigmoid(x):
    return 1 / (1+math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(-50)
```

# Tanh (1 to -1)

```python
def tanh(x):
    return (math.exp(x)-math.exp(-x))/(math.exp(x) + math.exp(-x))
```

```python
tanh(500)
```

```python
tanh(-500)
```

```python
tanh(50)
```

```python
tanh(0.500)
```

```python
tanh(-0.920)
```

# Relu

```python
def relu(x):
    return max(0,x)
```

```python
relu(100)
```

```python
relu(-500)
```

# Leaky Relu

```python
def leaky_relu(x):
    return max(0.01*x,x)
```

```python
leaky_relu(-100)
```

```python
leaky_relu(10)
```
